## Which Player is dismissed maximum by R Ashwin and How?

```python
Ashwin_df= df[df['bowler']=='R Ashwin']   #filtering from a data frame
Ashwin_df
```

```python
Player_df= Ashwin_df.groupby('player_dismissed')['player_dismissed'].count().idxmax()
Player_df
```

```python
Player_df= Ashwin_df.groupby('player_dismissed')['dismissal_kind'].value_counts().idxmax(axis=0,skipna=True)
Player_df
```
