#### 1) Drop the rows team1 or team2 has Pune,Delhi,Punjab. use contains() and dl_applied column

```python
df2=df1.drop([0])   #0 in this case is the label name. we have to drop dl_applied=0.
df2.head(5)
```

#### Method 1: To drop column value which has the values by join()

```python
Row_String_Values=['Pune','Delhi','Punjab']
df2[~df2.team1.str.contains('|'.join(Row_String_Values))]
```

#### Method 2: To Drop the str value of team1 column using contains | symbol 

```python
df2[~df2.team1.str.contains('Pune|Delhi|Punjab')]
```

```python
# when df [] is not defined it retuens boolean
~df2.team1.str.contains('Pune|Delhi|Punjab')
```

#### Method 3: Way to drop two columns based on the same conditions without using concat()

```python
df2[~df2['team1'].str.contains("Pune|Kochi|Gujarat") | ~df2['team2'].str.contains('Pune|Kochi|Gujarat')]
```

### 2. Fill the missing values of umpire 3 with the mode of the column

### Learning: if string is present in column, use mode() and give [] value to fill in all the NAN values. example, if i give [1] it replaces mode value of index 1 with every NAN value

```python
import pandas as pd
df_col=pd.read_csv('C:\\Users\\abc\\Desktop\\matches.csv',usecols=['umpire3'])
df_col.head(10)
```

#### We can directly assign a string of our choice in the column which has nan values in the column

```python
#df_col['umpire3'].fillna("Hi",inplace=True)
```

#### fill mode of umpire3 in umpire3 column. here, [1] represents the index value(nitin menon) which will be passed to the entire 'umpire3' column with nan values

```python
df_col['umpire3'].fillna(df_col['umpire3'].mode()[1],inplace=True)
```

### 3) Who has taken most number of wickets in a match when chasing(pick innings 2 &4) and print the umpire1 for that match

#### We will have to compare 2 data set for this question. and map with the match id obtainined which is common in both data sets

```python
import pandas as pd
df121=pd.read_csv('C:\\Users\\abc\\Desktop\\deliveries.csv')
df121.head(5)
```

```python
df_chase=df121[(df121['inning']%2==0)&df121['player_dismissed'].notnull()]   # filtering innings as 2&4 and filter notna players from player dismissed column
```

```python
df_bowler=df_chase.groupby('match_id')['bowler'].value_counts().idxmax()[0]   #[0] will five only the index number
```

```python
df_bowler
```

#### Mapping 'Match id' of deliveries data set to 'id 'of matches dataset

```python
import pandas as pd
df221=pd.read_csv('C:\\Users\\abc\\Desktop\\matches.csv')
df221.head()
```

```python
print(df221[df221['id']==df_bowler]['umpire1'])
```
